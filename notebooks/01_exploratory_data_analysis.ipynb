```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
```

```python
# Visualizing the df

DATA_PATH = "../data/raw/creditcard.csv"

df = pd.read_csv(DATA_PATH)

df.head()


# Understanding the size of the dataset

print(f"Linhas: {df.shape[0]}")
print(f"Colunas: {df.shape[1]}")

df.columns
# The Time column represents the time in seconds since the first transaction in the dataset.
# The Amount column represents the transaction amount.
# The Class column is the target variable, where 0 indicates a legitimate transaction and 1 indicates a fraudulent transaction.
# The features V1, V2, ..., V28 are the results of a PCA transformation applied to the original features of the dataset, which were removed for privacy and confidentiality reasons.

df.info()


df_describe = df.describe()
print(df_describe)

df.isnull().sum().sum()
# All items in the dataset are numeric, and there are no null values. The Class column is the target variable.

df.duplicated().sum()
```

```python
# Checking the number of duplicate transactions
df.duplicated().sum()

# 1081 duplicate transactions were found in the dataset.
```

```python
# Analyzing the Class column
df["Class"].value_counts()

""" Class
0    284315
1       492 """

# Visualizing the distribution of frauds *Very little amount of frauds*

sns.countplot(x = "Class" , data = df)

plt.title("Distribution of classes")
plt.show()

# Class imbalance

class_percentages = df["Class"].value_counts(normalize=True) * 100
class_percentages

# Accuracy is not a good metric for this dataset, as the dataset is highly imbalanced. 
# A model that predicts all transactions as legitimate would have an accuracy of 99.83%, but it would not be able to detect any fraudulent transactions.
```
