## 4. Target Variable Analysis

The `Class` feature represents whether a transaction is legitimate (0) or fraudulent (1). 

Analyze the distribution of fraudulent and legitimate transactions.

```python
df["Class"].value_counts()
```

```python
class_counts = df["Class"].value_counts()
class_percentages = df["Class"].value_counts(normalize=True) * 100

print("Class Counts:")
print(f"Legitimate: {class_counts[0]}")
print(f"Fraudulent: {class_counts[1]}")

print("\nClass Percentages:")
print(f"Legitimate: {class_percentages[0]:.2f}%")
print(f"Fraudulent: {class_percentages[1]:.2f}%")
```

```python
sns.countplot(data=df, x="Class", hue="Class", palette=["blue", "red"])
sns.set_theme(style="whitegrid", context="talk")

plt.title("Distribution of Transaction Classes")
plt.xlabel("Transaction Class")
plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=["Legitimate", "Fraudulent"])
plt.show()
```

### 4.1 Conclusion

The dataset is highly imbalanced.

- Legitimate transactions: 99.83%
- Fraudulent transactions: 0.17%

Because of this severe class imbalance, accuracy alone is not an appropriate evaluation metric. Future models will be evaluated using Precision, Recall, F1-score and PR-AUC.

## 5. Transaction Amount Analysis

The `Amount` feature represents the monetary value of each transaction.

This section explores the distribution of transaction amounts, identifies potential outliers, and compares transaction values between legitimate and fraudulent transactions.

```python
plt.figure(figsize=(10,5))

sns.histplot(df["Amount"], bins=50)

plt.title("Distribution of Transaction Amount")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(10,2))

sns.boxplot(x=df["Amount"])

plt.title("Boxplot of Transaction Amount")
plt.xlabel("Transaction Amount")

plt.show()
```

The Amount feature exhibits a strong positive (right) skew, with most transaction amounts concentrated at low values and numerous high-value outliers extending the distribution to the right.

```python
df.groupby("Class")["Amount"].describe()
```

```python
plt.Figure(figsize=(10,2))
sns.boxplot(data=df, x="Class", y="Amount")
plt.title("Boxplot of Transaction Amount by Class")
```

```python
fraud = df[df["Class"] == 1]

fraud["Amount"].max()
```

###  5.1 Conclusions

- The `Amount` feature exhibits a highly right-skewed distribution.
- Most transactions involve relatively small amounts, while a small number of transactions correspond to high-value purchases, resulting in numerous outliers.
- In this dataset, fraudulent transactions are concentrated in lower transaction amounts, with no fraudulent transaction exceeding of $2125,87.
- Although `Amount` provides useful information, it is unlikely to be sufficient on its own to distinguish fraudulent from legitimate transactions.

## 6. Time Analysis

The `Time` feature represents the elapsed time, in **seconds**, between each transaction and the first transaction recorded in the dataset.

This section aims to explore the temporal distribution of transactions and investigate whether fraudulent transactions exhibit any noticeable temporal patterns.

```python
df["Time"].describe()
```

```python

max_time = df["Time"].max()
max_time_hours = max_time / 3600
print(f"Maximum time in hours: {max_time_hours:.2f} hours")
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["Time"],
    bins=50,
    color="blue",
    edgecolor="white"

)

plt.title("Distribution of Transaction Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency")

plt.show()
```

```python
df.groupby("Class")["Time"].describe()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    data=df,
    x="Time",
    hue="Class",
    bins=50,
    element="step"
)

plt.title("Transaction Time by Class")

plt.show()
```

```python
fraud = df[df["Class"] == 1]

fraud["Time"].describe()
```

### 6.1 Conclusions

- The dataset covers approximately 48 hours of transaction data.
- Transactions are not uniformly distributed over time.
- Fraudulent transactions occur throughout the observation period.
- No clear temporal pattern is observed that would allow the `Time` feature alone to distinguish fraudulent from legitimate transactions.

## 7. Correlation Analysis

Correlation analysis helps identify the strength and direction of the linear relationship between variables.

In this section, we investigate how the features relate to one another and identify which variables are most strongly associated with fraudulent transactions.

### Correlation Matrix

The first step is to compute the correlation matrix, which contains the pairwise correlation coefficients between all numerical features in the dataset.

```python
correlation_matrix = df.corr()

correlation_matrix
```

### Correlation Heatmap

Since the correlation matrix contains many values, a heatmap provides a clearer visual representation of the relationships between variables.

```python
plt.figure(figsize=(18,14))

sns.heatmap(
    correlation_matrix,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation Matrix")

plt.show()
```

**Insights**

- Most feature pairs exhibit weak linear correlations.
- The main diagonal contains correlation values equal to 1, as each feature is perfectly correlated with itself.
- The `Amount` feature shows little linear correlation with the target variable (`Class`).
- The heatmap suggests that some PCA components may be more correlated with fraud than others; however, a more detailed analysis of the correlation coefficients is required.

### Correlation with the Target Variable

To better understand which features may be useful for fraud detection, we analyze the correlation of each feature with the target variable (`Class`).

```python
correlation_with_target = correlation_matrix["Class"]

correlation_with_target.sort_values()
```

### Feature Correlation with Fraud

The following chart ranks all features according to their correlation with the target variable.

```python
plt.figure(figsize=(8,10))

correlation_with_target.sort_values().plot(kind="barh", colormap="coolwarm")

plt.title("Feature Correlation with Target")

plt.show()
```

**Insights**

- Some variables exhibit stronger positive or negative correlations with fraud.
- Features with correlations close to zero are less informative when considered individually.
- Both positive and negative correlations can provide useful information for machine learning models.

### 7.1 Conclusions

- Most features exhibit weak linear correlations with one another.
- When considering the target variable (`Class`), `V11` shows the strongest positive correlation, while `V17` presents the strongest negative correlation.
- These features may provide valuable information for distinguishing fraudulent transactions and are likely to be useful during model training.
- Although correlation does not imply causation, the observed relationships suggest that higher values of `V11` are associated with a greater likelihood of fraud, whereas lower values of `V17` are also associated with a greater likelihood of fraud.
